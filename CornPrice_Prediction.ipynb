# Original Data(1986-01 ~ 2022-03) 분석

````python
import numpy as np
import pandas as pd

data = pd.read_excel('/content/drive/MyDrive/CORNdataset.xlsx', index_col=0)
data
````

````python
#ADF 검정 결과와 함께 해당 변수의 시계열 그래프를 출력합니다.

from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt

def adf_test(series, title):
    result = adfuller(series.dropna())
    print(f'Column: {title}')
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    print(f'Critical Values:')
    for key, value in result[4].items():
        print(f'   {key}: {value}')
    
    # plot time series
    fig, ax = plt.subplots(figsize=(12,6))
    ax.plot(series.index, series.values)
    ax.set_title(title)
    ax.set_xlabel('Time')
    ax.set_ylabel(title)
    plt.show()

for col in data.columns:
    adf_test(data[col], col)
````

#ARIMAX


ADF 검정에서 귀무가설은 "해당 시계열 데이터에 단위근(unit root)이 존재한다" 입니다. 대립가설은 "해당 시계열 데이터에 단위근이 존재하지 않는다"입니다.

ADF 검정의 결과 p-value가 0.1 이상이면 귀무가설을 기각할 수 없으므로, 해당 시계열 데이터는 단위근이 존재한다고 할 수 있습니다. 즉, 해당 시계열 데이터는 비정상(non-stationary)이며, 추세나 계절성 등의 요인이 포함되어 있을 가능성이 높습니다.

따라서, 시계열 데이터를 분석하거나 모델링할 때, 비정상 시계열 데이터를 그대로 사용하는 것은 부적절할 수 있습니다. 비정상 시계열 데이터를 정상화(stationarize)시켜 분석하거나 모델링하는 것이 좋습니다.


p-value가 0.1 이상 : SOI,WTI종가,금리,옥수수선물

p-value가 0.1 이하 : NonComm_Positions_Long_All, 달러화가치


변수간 스케일 조정효과를 누리기 위해서 Yeo-Johnson 변환을 수행한다.

VECM은 공적분(cointegrated)된 시계열 변수들의 장기 평형관계(long-term equilibrium relationship)와 짧은 기간의 플러그메인(fluctuations)을 분석할 수 있는 모델입니다. 공적분(cointegration) 관계가 있는 시계열 변수들은 장기적인 평형관계를 가지기 때문에 단변량(non-stationary) 시계열 변수들과 달리 차분이나 변환을 하지 않아도 됩니다. 따라서 VECM을 적용할 때에는 굳이 차분이나 여존슨(Johansen) 변환을 수행할 필요는 없습니다. 다만, 공적분(cointegration) 관계를 가지는지 여부를 확인하기 위해 여존슨(Johansen) 검정을 수행하는 것이 일반적입니다.


````python
import pandas as pd
from statsmodels.tsa.api import VAR

results = []
for i in range(1, 13):
    model = VAR(data)
    results.append(model.fit(i).aic)
    results.append(model.fit(i).fpe)
    results.append(model.fit(i).hqic)
    results.append(model.fit(i).bic)

results = np.array(results).reshape(12, 4)
results_df = pd.DataFrame(results, index=range(1, 13), columns=['AIC', 'FPE', 'HQ', 'SC'])

print(results_df)
````

최적의 시차는 AIC, FPE, HQ, SC 값 중에서 가장 작은 값을 갖는 시차입니다. 따라서 위의 결과를 보면 2번째 row의 값이 AIC, FPE, HQ, SC 중에서 가장 작습니다. 따라서 최적의 시차는 2입니다.

````python
from statsmodels.tsa.vector_ar.vecm import coint_johansen

data_t = data[['NonComm_Positions_Long_All', 'WTI종가', '달러화가치', '옥수수선물']]
jres = coint_johansen(data, det_order=0, k_ar_diff=2)

print('Test for the cointegration(trace)')
print('No. of cointegration Test 10% 5% 1%')
for i in range(len(data_t.columns)):
    print('r <=', i, jres.lr1[i], jres.cvt[i][0], jres.cvt[i][1], jres.cvt[i][2])
````

 "r <= 0"에서 검정 통계량은 약198.212입니다. 이 값은 10% 유의수준에서 임계값인 91.109보다 크고, 5% 유의수준에서의 임계값인 95.754보다 크고, 1% 유의수준에서의 임계값인 104.963보다 큽니다. 따라서 "r <= 0"의 귀무가설을 기각할 수 있으며, 변수들 간에는 적어도 1개의 공적분 관계가 존재한다고 할 수 있습니다.

마찬가지로 "r <= 1", "r <= 2", "r <= 3"에 대해서도 유사한 방식으로 해석할 수 있습니다. 예를 들어, "r <= 1"의 경우 검정 통계량은 약106.468이며, 10% 유의수준에서의 임계값인 65.820보다 크고, 5% 유의수준에서의 임계값인 69.818보다 크고, 1% 유의수준에서의 임계값인 77.820보다 큽니다. 따라서 "r <= 1"의 귀무가설을 기각할 수 있으며, 변수들 간에는 적어도 2개의 공적분 관계가 존재한다고 할 수 있습니다.

#LSTM

다음은 lag가2, epochs=500, batch_size=20,units=10 인 LSTM에 실험데이터셋을 학습하고 예측한 결과입니다.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM

# 데이터 불러오기
df = pd.read_excel('/content/drive/MyDrive/CornTestData.xlsx', index_col=0)

# 훈련 데이터와 테스트 데이터로 분리
train_start, train_end = '2002-01', '2017-12'
test_start, test_end = '2018-01', '2021-12'
train = df.loc[train_start:train_end, :]
test = df.loc[test_start:test_end, :]

# 종속변수와 독립변수 나누기
X_train = train[['NonComm_Positions_Long_All', 'WTI종가', '달러화가치']]
y_train = train[['옥수수선물']]
X_test = test[['NonComm_Positions_Long_All', 'WTI종가', '달러화가치']]
y_test = test[['옥수수선물']]

# 스케일링
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)
y_train_scaled = scaler_y.fit_transform(y_train)
y_test_scaled = scaler_y.transform(y_test)

# 데이터 전처리
def create_dataset(X, y, time_steps):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:i+time_steps, :])
        ys.append(y[i+time_steps, 0])
    return np.array(Xs), np.array(ys)

time_steps=2

X_train_processed, y_train_processed = create_dataset(X_train_scaled, y_train_scaled, time_steps)
X_test_processed, y_test_processed = create_dataset(X_test_scaled, y_test_scaled, time_steps)

# 모델 학습
model = Sequential()
model.add(LSTM(units=10, input_shape=(time_steps, X_train_processed.shape[2])))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')
history = model.fit(X_train_processed, y_train_processed, epochs=500, batch_size=20, verbose=1)

# 예측
y_train_predicted = model.predict(X_train_processed)
y_test_predicted = model.predict(X_test_processed)

# 스케일링 복원
y_train_predicted = scaler_y.inverse_transform(y_train_predicted)
y_test_predicted = scaler_y.inverse_transform(y_test_predicted)
y_train_actual = scaler_y.inverse_transform(y_train_processed.reshape(-1, 1))
y_test_actual = scaler_y.inverse_transform(y_test_processed.reshape(-1, 1))

# MAPE 계산
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

train_mape = mape(y_train_actual, y_train_predicted)
test_mape = mape(y_test_actual, y_test_predicted)
print('Train MAPE:', train_mape)
print('Test MAPE:', test_mape)
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM

# 데이터 불러오기
df = pd.read_excel('/content/drive/MyDrive/CornTestData.xlsx', index_col=0)

# 훈련 데이터와 테스트 데이터로 분리
train_start, train_end = '2002-01', '2017-12'
test_start, test_end = '2018-01', '2021-12'
train = df.loc[train_start:train_end, :]
test = df.loc[test_start:test_end, :]

# 종속변수와 독립변수 나누기
X_train = train[['WTI종가']]
y_train = train[['옥수수선물']]
X_test = test[['WTI종가']]
y_test = test[['옥수수선물']]

# 스케일링
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)
y_train_scaled = scaler_y.fit_transform(y_train)
y_test_scaled = scaler_y.transform(y_test)

# 데이터 전처리
def create_dataset(X, y, time_steps):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:i+time_steps, :])
        ys.append(y[i+time_steps, 0])
    return np.array(Xs), np.array(ys)

time_steps=2

X_train_processed, y_train_processed = create_dataset(X_train_scaled, y_train_scaled, time_steps)
X_test_processed, y_test_processed = create_dataset(X_test_scaled, y_test_scaled, time_steps)

# 모델 학습
model = Sequential()
model.add(LSTM(units=10, input_shape=(time_steps, X_train_processed.shape[2])))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')
history = model.fit(X_train_processed, y_train_processed, epochs=500, batch_size=20, verbose=0)

# 예측
y_train_predicted = model.predict(X_train_processed)
y_test_predicted = model.predict(X_test_processed)

# 스케일링 복원
y_train_predicted = scaler_y.inverse_transform(y_train_predicted)
y_test_predicted = scaler_y.inverse_transform(y_test_predicted)
y_train_actual = scaler_y.inverse_transform(y_train_processed.reshape(-1, 1))
y_test_actual = scaler_y.inverse_transform(y_test_processed.reshape(-1, 1))

# MAPE 계산
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

train_mape = mape(y_train_actual, y_train_predicted)
test_mape = mape(y_test_actual, y_test_predicted)
print('Train MAPE:', train_mape)
print('Test MAPE:', test_mape)
````
